## Task 1 — Load and Inspect the Data (20 pts)

```python
# Import required libraries
import pandas as pd
import numpy as np

# Load the Titanic CSV into a DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

print("Dataset loaded successfully!")
```

```python
# Display the first 10 rows
df.head(10)
```

```python
# Print the shape of the dataset (rows, columns)
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# List all column names
print("Column names:")
print(df.columns.tolist())
```

---
## Task 2 — Clean the Data (20 pts)

```python
# Check for missing values in each column
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Drop rows where the Age column is missing
df_clean = df.dropna(subset=['Age'])

print(f"Original dataset size: {len(df)} rows")
print(f"After dropping missing Age values: {len(df_clean)} rows")
print(f"Rows removed: {len(df) - len(df_clean)}")
```

```python
# Confirm no missing Age values remain
print(f"Missing Age values after cleaning: {df_clean['Age'].isnull().sum()}")
```

---
## Task 3 — Filter and Sort (30 pts)

```python
# Create a new DataFrame with only passengers who survived (Survived == 1)
survivors = df_clean[df_clean['Survived'] == 1]

print(f"Number of survivors: {len(survivors)}")
```

```python
# From the filtered DataFrame, show only passengers older than 30
survivors_over_30 = survivors[survivors['Age'] > 30]

print(f"Survivors older than 30: {len(survivors_over_30)}")
```

```python
# Sort the result by Age in descending order and display the top 10 rows
survivors_sorted = survivors_over_30.sort_values(by='Age', ascending=False)

print("Top 10 oldest survivors (over 30):")
survivors_sorted.head(10)
```

---
## Task 4 — Summary Statistics (30 pts)

```python
# Calculate the mean, median, and standard deviation of the Age column
age_mean = df_clean['Age'].mean()
age_median = df_clean['Age'].median()
age_std = df_clean['Age'].std()

print("Age Statistics:")
print(f"  Mean: {age_mean:.2f}")
print(f"  Median: {age_median:.2f}")
print(f"  Standard Deviation: {age_std:.2f}")
```

```python
# Find the average fare paid by passengers in each class (Pclass)
avg_fare_by_class = df_clean.groupby('Pclass')['Fare'].mean()

print("Average Fare by Passenger Class:")
for pclass, fare in avg_fare_by_class.items():
    print(f"  Class {pclass}: ${fare:.2f}")
```

```python
# Print the total number of survivors vs non-survivors
survival_counts = df_clean['Survived'].value_counts()

print("Survival Counts:")
print(f"  Survivors: {survival_counts[1]}")
print(f"  Non-survivors: {survival_counts[0]}")
print(f"  Survival Rate: {survival_counts[1] / len(df_clean) * 100:.1f}%")
```
